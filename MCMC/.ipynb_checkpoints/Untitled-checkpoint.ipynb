```python
def logp(theta, sealevel, deltat, temperatures, sigma=1):
    N = len(sealevel)
    alpha, Teq, S0 = theta[0], theta[1], theta[2]
    model = [S_i + deltat * alpha * (temperatures[i]-Teq) for i in range(0, len(temperatures))]
    resid = sealevel - model
    log_likelihood = -N / 2 * math.log(
        2 * math.pi) - N * math.log(sigma) - 1 / 2 * sum(resid**2) / sigma**2
    log_prior = 0  # log(1)
    log_posterior = log_likelihood + log_prior
    return log_posterior
```

```python
def chain(parameters, temperatures, deltat, N=30000):
    alpha, Teq, S0 = parameters[0], parameters[1], parameters[2]
    theta = [S0, alpha, Teq]
    print('Random estimate for parameters -', theta)

    lp = logp(theta, time, y)
    theta_best = theta
    lp_max = lp
    theta_new = [0] * 2
    accepts = 0
    mcmc_chains = [[0., 0., 0.]] * N
    mcmc_chains = np.array(mcmc_chains)
    step = np.array([[.1,0,0], [0,.1,0], [0,0,.1]])
    # Defining variables for Adaptive metropolis Hastings -
    sd = 2.38**2 / 2

    for i in range(N):
        theta_new = list(np.random.multivariate_normal(theta, step))
        lp_new = logp(theta, sealevel, deltat, temperatures, sigma=1)
        lq = lp_new - lp
        lr = math.log(np.random.uniform(0, 1))
        if (lr < lq):
            theta = theta_new
            lp = lp_new
            accepts += 1
            if lp > lp_max:
                theta_best = theta
                lp_max = lp
        mcmc_chains[i] = theta
    return mcmc_chains
```
